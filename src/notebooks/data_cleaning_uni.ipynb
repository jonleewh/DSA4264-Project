# Combined Uni Data

## Merging of University Data

```python
import pandas as pd
import re

# --- NUS ---
nus_df = filtered_nus_df.copy()
nus_df["department"] = nus_df["department"]   # using department now
nus_df["code"] = nus_df["moduleCode"]
nus_df["university"] = "NUS"

nus_df = nus_df[["code", "title", "department", "description", "university"]]


# --- NTU ---
ntu_df = filtered_ntu_df.copy()
ntu_df["department"] = ntu_df["department"]
ntu_df["university"] = "NTU"

ntu_df = ntu_df[["code", "title", "department", "description", "university"]]


# --- SUTD ---
sutd_df = filtered_sutd_df.copy()

# fix list issue
sutd_df["department"] = sutd_df["department"].apply(
    lambda x: x[0] if isinstance(x, list) else x
)

sutd_df["university"] = "SUTD"

sutd_df = sutd_df[["code", "title", "department", "description", "university"]]


# --- Combine ---
combined_df = pd.concat([nus_df, ntu_df, sutd_df], ignore_index=True)


# --- Clean illegal characters ---
illegal_chars_re = re.compile(r"[\x00-\x08\x0B-\x0C\x0E-\x1F]")

combined_df = combined_df.applymap(
    lambda x: illegal_chars_re.sub(" ", x) if isinstance(x, str) else x
)

# strip whitespace
combined_df["department"] = combined_df["department"].str.strip()


# --- Add serial number ---
combined_df.insert(0, "s/n", range(1, len(combined_df) + 1))


combined_df.head()
```

## Extracting relevant skills from course description

```python
import re
import pandas as pd
from sentence_transformers import SentenceTransformer, util

# -------------------------
# 1. NORMALISE SPELLING
# -------------------------
def normalize_spelling(text):
    if not isinstance(text, str):
        return text

    replacements = {
        "optimisation": "optimization",
        "organisation": "organization",
        "behaviour": "behavior",
        "labour": "labor",
        "analyse": "analyze",
        "analysing": "analyzing",
        "modelling": "modeling",
        "colour": "color",
        "favour": "favor",
        "centre": "center",
        "theatre": "theater",
        "programme": "program"
    }

    for k, v in replacements.items():
        text = text.replace(k, v)

    return text


# -------------------------
# 2. CLEAN TEXT
# -------------------------
def clean_text(text):
    if pd.isna(text):
        return ""

    text = str(text).lower()
    text = normalize_spelling(text)

    text = re.sub(r"[\x00-\x08\x0B-\x0C\x0E-\x1F]", " ", text)
    text = re.sub(r"\s+", " ", text)

    return text.strip()


combined_df["description_clean"] = combined_df["description"].apply(clean_text)


# -------------------------
# 3. SPLIT INTO CHUNKS
# -------------------------
def split_into_chunks(text):
    if not isinstance(text, str) or not text.strip():
        return []

    chunks = re.split(r"[.;:]\s+|\.\s+|\n+", text)
    return [c.strip() for c in chunks if c.strip()]


combined_df["description_chunks"] = combined_df["description_clean"].apply(split_into_chunks)


# -------------------------
# 4. NORMALISE SKILL LIST
# -------------------------
def normalize_skill_list(skills):
    return [normalize_spelling(s.lower()) for s in skills]


hard_skills = normalize_skill_list([
    # computing / quantitative
    "python", "sql", "java", "c", "c++", "programming",
    "machine learning", "deep learning", "natural language processing",
    "data analysis", "data visualization", "statistics", "probability",
    "linear algebra", "optimization", "regression analysis",
    "database systems", "computer vision", "cloud computing", "cybersecurity",
    "game theory", "simulation modeling", "modeling",

    # business / finance
    "accounting", "managerial accounting", "financial accounting",
    "auditing", "audit planning", "risk assessment", "internal controls",
    "financial reporting", "financial statement analysis",
    "corporate governance", "strategy formulation", "strategy implementation",
    "industry analysis", "portfolio management", "investment banking",
    "financial planning", "marketing planning", "marketing strategy",
    "project management", "budgeting", "capital structure",
    "financial markets", "market efficiency", "credit management",

    # arts / humanities / social sciences
    "visual analysis", "comparative analysis", "close reading",
    "literary analysis", "textual analysis", "historical analysis",
    "critical analysis", "archival research", "curatorial practice",
    "curatorial methodologies", "exhibition studies", "art historical methods",
    "ethnography", "field ethnography", "fieldwork", "cultural analysis",
    "social analysis", "policy analysis", "translation", "translation practice",
    "rhetorical analysis", "discourse analysis", "phonological analysis",
    "textual criticism", "comparative study", "source analysis",
    "interdisciplinary analysis", "historical interpretation",

    # architecture / design / urban
    "architectural design", "landscape design", "urban design",
    "architectural history", "architectural theory", "urban planning",
    "heritage conservation", "heritage management", "site analysis",
    "mapping", "architectural lighting", "architectural acoustics",
    "passive design", "environmentally responsive design",
    "climate responsive design", "ecological design",
    "structural mechanics", "structural analysis", "building systems",
    "building services systems", "energy efficiency", "material analysis",

    # biomedical / engineering
    "signal processing", "medical instrumentation", "thermodynamic analysis",
    "kinetic analysis", "biomechanics", "tissue engineering",
    "drug delivery", "microfabrication", "biosensors",
    "genetic engineering", "computational modeling",
    "control theory", "robot kinematics", "motion tracking",
    "finite element analysis", "wastewater treatment",
    "transport infrastructure design", "stability assessment",
    "groundwater seepage analysis", "structural stability",

    # performing arts / media
    "performance skills", "dance performance", "music performance",
    "choreography", "news writing", "media production"
])

soft_skills = normalize_skill_list([
    "communication", "leadership", "teamwork", "collaboration",
    "critical thinking", "problem solving", "decision-making",
    "strategic thinking", "professional conduct", "ethics",
    "ethical responsibility", "creativity", "adaptability",
    "analytical skills", "presentation skills", "research skills",
    "writing skills", "close reading", "interpretation",
    "fieldwork skills", "curatorial practice", "stylistic sensitivity"
])

all_skills = hard_skills + soft_skills
hard_skill_set = set(hard_skills)
soft_skill_set = set(soft_skills)


# -------------------------
# 5. EMBEDDING MODEL
# -------------------------
model = SentenceTransformer("all-MiniLM-L6-v2")
skill_embeddings = model.encode(all_skills, convert_to_tensor=True)


# -------------------------
# 6. CHUNK MATCHING
# -------------------------
def extract_skills_by_chunks(chunks, top_k=5, min_score=0.30):
    if not chunks:
        return []

    chunk_embeddings = model.encode(chunks, convert_to_tensor=True)
    sim_matrix = util.cos_sim(chunk_embeddings, skill_embeddings)

    max_scores = sim_matrix.max(dim=0).values

    scored_skills = [
        (all_skills[i], float(max_scores[i]))
        for i in range(len(all_skills))
        if float(max_scores[i]) >= min_score
    ]

    scored_skills = sorted(scored_skills, key=lambda x: x[1], reverse=True)[:top_k]

    return [skill for skill, score in scored_skills]


combined_df["skills_embedding"] = combined_df["description_chunks"].apply(
    lambda chunks: extract_skills_by_chunks(chunks)
)

combined_df["hard_skills"] = combined_df["skills_embedding"].apply(
    lambda skills: [s for s in skills if s in hard_skill_set]
)

combined_df["soft_skills"] = combined_df["skills_embedding"].apply(
    lambda skills: [s for s in skills if s in soft_skill_set]
)


# -------------------------
# 7. CHECK EMPTY
# -------------------------
empty_rows = combined_df[
    (combined_df["hard_skills"].str.len().fillna(0) == 0) &
    (combined_df["soft_skills"].str.len().fillna(0) == 0)
]

print(f"Empty rows: {len(empty_rows)} / {len(combined_df)}")
print(f"Percentage empty: {len(empty_rows)/len(combined_df):.2%}")
```

```python
# combined_df.to_excel("../../data/reupdated_skills.xlsx", index=False)

combined_df["num_skills"] = combined_df["skills_embedding"].apply(len)
combined_df["num_hard_skills"] = combined_df["hard_skills"].apply(len)
combined_df["num_soft_skills"] = combined_df["soft_skills"].apply(len)

print("Avg total skills:", combined_df["num_skills"].mean())
print("Avg hard skills:", combined_df["num_hard_skills"].mean())
print("Avg soft skills:", combined_df["num_soft_skills"].mean())

print("Median total skills:", combined_df["num_skills"].median())
print("Median hard skills:", combined_df["num_hard_skills"].median())
print("Median soft skills:", combined_df["num_soft_skills"].median())

print(combined_df["num_skills"].value_counts().sort_index())
```
